```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/P1_Credit_Risk_Scoring.csv")

print("Dataset Information:")
df.info()

print("\nMissing values:")
print(df.isnull().sum())

print("\nNumber of duplicate rows:")
duplicate_rows = df.duplicated().sum()
print(duplicate_rows)

if duplicate_rows > 0:
    df.drop_duplicates(inplace=True)
    print(f"\n{duplicate_rows} duplicate rows removed. New shape: {df.shape}")
else:
    print("\nNo duplicate rows found.")

df
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Approved', data=df, palette='viridis')
plt.title('Distribution of Loan Approval Status')
plt.xlabel('Approved (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

print("Correlation Matrix:")
display(correlation_matrix)


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
print("Correlation of 'Approved' with other variables:")
display(correlation_matrix['Approved'].sort_values(ascending=False))
```

From the correlation values, we can observe:

*   **Negative Correlation with Past_Defaults:** As expected, there's a strong negative correlation (-0.41) between `Approved` and `Past_Defaults`. This means that applicants with more past defaults are less likely to be approved.
*   **Negative Correlation with Existing_Loans_Count:** There's also a negative correlation (-0.14) with `Existing_Loans_Count`, suggesting that a higher number of existing loans might slightly decrease the chances of approval.
*   **Positive Correlation with Credit_History_Length_Years:** A positive correlation (0.13) with `Credit_History_Length_Years` indicates that a longer credit history might slightly increase the likelihood of approval.
*   **Weak Correlations:** Other features like `Age`, `Income_Monthly`, and `Loan_Amount_Requested` show very weak correlations with `Approved` status, indicating they might not be strong linear predictors on their own.
